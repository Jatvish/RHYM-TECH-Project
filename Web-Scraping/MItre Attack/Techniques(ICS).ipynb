```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://attack.mitre.org/techniques/ics/'
page = requests.get(url)
```

```python
if page.status_code == 200:
    soup = BeautifulSoup(page.text,'html')
else:
    print("request failed")
```

```python
table_body = soup.find("tbody")

result = []
for table_row in table_body.findAll("tr"):
    table_cols = table_row.findAll("td")
    result.append({"id":table_cols[0].text.replace("\n", "").strip(), "name":table_cols[1].text.replace("\n", "").strip(),"Description":table_cols[2].text.replace("\n", "").strip()})
```
