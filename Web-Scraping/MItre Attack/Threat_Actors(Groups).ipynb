```python
from bs4 import BeautifulSoup
import requests
```

```python
url =  'https://attack.mitre.org/groups/'
page = requests.get(url)
```

```python
if page.status_code == 200:
    soup = BeautifulSoup(page.text, 'html')
else:
    print("request failed")
```

```python
table_data = soup.find("table")

table_body = table_data.find("tbody")

result = []
for table_row in table_body.findAll("tr"):
    table_cols = table_row.findAll("td")
    result.append({"id":table_cols[0].text.replace("\n", "").strip(), "name":table_cols[1].text.replace("\n", "").strip(),"Associated Groups":table_cols[2].text.replace("\n", "").strip(),"Description":table_cols[3].text.replace("\n", "").strip()})


result
```

```python
all_links = []

for table_row in table_body.findAll("tr"):
    table_col = table_row.findAll("td")
    links_in_col = table_col[1].findAll("a")
    
    # Extract href attribute
    for link in links_in_col:
        href = link.get("href")
        if href:
            all_links.append(href)


print(all_links)
# print(all_links[0])
```

```python
def fetching_meta_information(soup2):
    result2 = []
    tname_list = []

    try:
        # card_body = soup2.find("div", class_="card-body").text
        
        table_data = soup2.find("table")
        table_body = table_data.find("tbody")
        # print(table_body)

        for table_row in table_body.findAll("tr"):
            table_cols = table_row.findAll("td")
            tname = table_cols[-2].text.strip()
            tname_list.append(tname)
            
        card_body = soup2.find("div", class_="card-body")      
        if card_body:
                divs = soup2.findAll('div', class_="row card-data")

                data = []
                for div in divs:
                    data.append(div.get_text(strip=True))
            
                # Ensure that data contains all required elements
                if len(data) >= 5:
                    result2.append({
                        "ID": data[0] if data[0] else "N/A",
                        "Contributers": data[1] if data[1] else "N/A",
                        "Version": data[2] if data[2] else "N/A",
                        "Created": data[3] if data[3] else "N/A",
                        "Last Modified": data[4] if data[4] else "N/A",
                        "Technique name": tname_list
                    })
                else:
                    result2.append({
                        "ID": data[0] if len(data) > 0 else "N/A",
                        "Contributers": data[1] if len(data) > 1 else "N/A",
                        "Version": data[2] if len(data) > 2 else "N/A",
                        "Created": data[3] if len(data) > 3 else "N/A",
                        "Last Modified": data[4] if len(data) > 4 else "N/A",
                        "Technique name": tname_list
                    })
        else:
             print("No card-body found")  
        
        
        
        
    except Exception as e:
        print(f"An error occurred: {e}")

    return(result2)
```

```python

for link in all_links:
    link = "https://attack.mitre.org"+link
    url2 = link
    page2 = requests.get(url2)
    

    if page2.status_code == 200:
        soup2 = BeautifulSoup(page2.text,'html')
        data = fetching_meta_information(soup2)
        print(data)
    else:
        print("request not found")
```
