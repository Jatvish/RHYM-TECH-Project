```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://malpedia.caad.fkie.fraunhofer.de/actors'
page = requests.get(url)
```

```python
if page.status_code == 200:
    soup = BeautifulSoup(page.text,'html')
else:
    print("request failed")
```

```python
table_data = soup.find("table")
```

```python
table_body = table_data.find("tbody")
```

```python
result = []
count = 0
for table_row in table_body.findAll("tr"):
    count = count +1
    table_cols = table_row.findAll("td")
    span_tag = table_row.find('span')
    if span_tag and span_tag.has_attr('title'):
        title_value = span_tag.get('title')
    else:
        title_value = 'N/A'  # Default value if not found
    
    result.append({"id":count, "name":table_cols[4].text.replace("\n", "").strip(),"Coverage":table_cols[5].text.replace("\n", "").strip(),"Country":title_value})
```

```python
all_links = []

for table_row in table_body.findAll("tr"):
    
    links = table_row.get('data-href')
    all_links.append(links)
    # print(links)
    



# print(all_links)
```

```python
def fetching_meta_information(soup2):
    result2 = []
    name_list = []
    desc_list = []
    af_list = []

    try:
        desc = soup2.find("p").text.strip()
        # print(desc)
        name = soup2.find("h2").text.replace("(Back to overview)","").replace("\xa0\n        ","").replace("\n        \n        ","")
        # print(name)
        div = soup2.find('div', {'style': 'margin:-3px;'})
        if div:
            af = div.findAll("span")
    
            for x in af:
                y = x.text.strip()
                af_list.append(y)
            # name_list.append(name)
            # desc_list.append(desc)
            # print(af_list)
            result2.append({"name":name,"associated families":af_list,"description":desc})
        print(result2)
        return(result2)
        
    except Exception as e:
        print(f"An error occurred: {e}")

    # print(name_list)
    # print(desc_list)
    
```

```python
import sqlite3

def insert_data_into_table1(result):
    # Connect to SQLite database (or create it if it doesn't exist)
    conn = sqlite3.connect('RHYM(database).db')
    cursor = conn.cursor()

    # Create the table if it doesn't exist
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS malpedia_ThreatActors1 (
            id TEXT,
            name TEXT,
            Coverage TEXT,
            Country TEXT
        )
    ''')

    # Insert data into the table
    for item in result:
        id = item['id']
        name = item['name']
        Coverage = item['Coverage']
        Country = item['Country']

        cursor.execute('''
            INSERT INTO malpedia_ThreatActors1 (id, name, Coverage, Country)
            VALUES (?, ?, ?, ?)
        ''', (id, name, Coverage, Country))

    # Commit the transaction and close the connection
    conn.commit()
    conn.close()
```
